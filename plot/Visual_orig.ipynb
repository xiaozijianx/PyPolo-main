```python
import pickle
import numpy as np
import pandas as pd
import os
from copy import deepcopy
from pathlib import Path
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors

import pypolo2
```

```python
# data.npz
grid_x = 20
grid_y = 20
Strategy = "AirPollutionexperimentforforestfire_Nonmyopic_Adaptive_Sensing_Operation"
sche_step = 8
adaptive_step = 2
Env = "Dynamic"
alpha = 0.5
adaptive = "Adaptive"
With_water = "Without_Replenish"
spray = "Control"
effect_threshold = 0.0
randomsource = False
team_size = 7
date = "2023072207080401"
filename = './outputs/x{}_y{}_{}_{}_{}_{}_teamsize{}_threshold{}_schestep{}_adaptivestep{}_result_{}_{}.npz'.format(grid_x, grid_y,\
    Strategy,Env,With_water,spray,team_size,effect_threshold,sche_step,adaptive_step,date,randomsource)
data = np.load(filename)
env_list_loaded1 = data['env_list']
# print(env_list_loaded1[0].shape)

mean_pollution_list = []
max_pollution_list = []
for env in env_list_loaded1:
    mean = np.mean(env)
    max = np.max(env)
    mean_pollution_list.append(mean)
    max_pollution_list.append(max)
mean_pollution1 = np.array(mean_pollution_list)
max_pollution1 = np.array(max_pollution_list)

Strategy = "AirPollutionexperimentforforestfire_Nonmyopic_Adaptive_Operation"
filename = './outputs/x{}_y{}_{}_{}_{}_{}_teamsize{}_threshold{}_schestep{}_adaptivestep{}_result_{}_{}.npz'.format(grid_x, grid_y,\
    Strategy,Env,With_water,spray,team_size,effect_threshold,sche_step,adaptive_step,date,randomsource)
data2 = np.load(filename)
env_list_loaded2 = data2['env_list']
# print(env_list_loaded1[0].shape)

mean_pollution_list = []
max_pollution_list = []
for env in env_list_loaded2:
    mean = np.mean(env)
    max = np.max(env)
    mean_pollution_list.append(mean)
    max_pollution_list.append(max)
mean_pollution2 = np.array(mean_pollution_list)
max_pollution2 = np.array(max_pollution_list)

Strategy = "AirPollutionexperimentforforestfire_Without_Operation"
filename = './outputs/x{}_y{}_{}_{}_{}_{}_teamsize{}_threshold{}_schestep{}_adaptivestep{}_result_{}_{}.npz'.format(grid_x, grid_y,\
    Strategy,Env,With_water,spray,team_size,effect_threshold,sche_step,adaptive_step,date,randomsource)
data3 = np.load(filename)
env_list_loaded3 = data3['env_list']
# print(env_list_loaded1[0].shape)

mean_pollution_list = []
max_pollution_list = []
for env in env_list_loaded3:
    mean = np.mean(env)
    max = np.max(env)
    mean_pollution_list.append(mean)
    max_pollution_list.append(max)
mean_pollution3 = np.array(mean_pollution_list)
max_pollution3 = np.array(max_pollution_list)

# Matplotlib
fig, ax = plt.subplots(1, 2, figsize=(8, 7))
ax1,ax2 = ax
# 
ax1.plot(mean_pollution1, label='Sensing_Operation')
ax1.plot(mean_pollution2, label='Operation')
ax1.plot(mean_pollution3, label='No_Control')
ax1.legend()
ax1.set_xlabel('time')
ax1.set_ylabel('Mean pollution')
ax1.set_title('Mean pollution Comparison')

# 
ax2.plot(max_pollution1, label='Sensing_Operation')
ax2.plot(max_pollution2, label='Operation')
ax2.plot(max_pollution3, label='No_Control')
ax2.legend()
ax2.set_xlabel('time')
ax2.set_ylabel('Max pollution')
ax2.set_title('Max pollution Comparison')
ax2.set_ylim(150, 500)

# # 
plt.show()
```
