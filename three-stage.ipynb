```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 模拟数据
data = [
    (0, 0, 0),  # 初始位置
    (1, 1, 0),  # 移动动作1
    (2, 1, 1),  # 移动动作2，洒水动作1
    (3, 2, 1),  # 移动动作3，洒水动作1
    (4, 3, 1),  # 移动动作4，洒水动作1
    (5, 4, 1),  # 移动动作5，洒水动作1
    (6, 4, 2),  # 移动动作6，洒水动作2
    (7, 4, 3)   # 移动动作7，洒水动作3
]

# 提取数据
time = [d[0] for d in data]
move_actions = [d[1] for d in data]
spray_actions = [d[2] for d in data]

# 创建三维图形
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制点
ax.scatter(time, move_actions, spray_actions, c='r', marker='o')

# 绘制原点到三个坐标轴的线
ax.plot([0, max(time)], [0, 0], [0, 0], 'k-')  # X轴
ax.plot([0, 0], [0, max(move_actions)], [0, 0], 'k-')  # Y轴
ax.plot([0, 0], [0, 0], [0, max(spray_actions)], 'k-')  # Z轴

# 添加坐标轴标注
ax.text(max(time), 0, 0, '时间', ha='left', va='bottom')
ax.text(0, max(move_actions), 0, '移动动作', ha='right', va='bottom')
ax.text(0, 0, max(spray_actions), '洒水动作', ha='left', va='top')

# 添加坐标轴刻度
ax.set_xticks([0, max(time)])
ax.set_yticks([0, max(move_actions)])
ax.set_zticks([0, max(spray_actions)])

# 反转Y轴和Z轴
ax.invert_yaxis()
ax.invert_zaxis()

# 绘制网格线
ax.grid(True)

# 设置坐标轴范围
ax.set_xlim([0, max(time)])
ax.set_ylim([max(move_actions), 0])
ax.set_zlim([max(spray_actions), 0])

# 绘制坐标轴箭头
ax.quiver(0, 0, 0, max(time), 0, 0, length=0.1, arrow_length_ratio=0.3, color='black')
ax.quiver(0, 0, 0, 0, max(move_actions), 0, length=0.1, arrow_length_ratio=0.3, color='black')
ax.quiver(0, 0, 0, 0, 0, max(spray_actions), length=0.1, arrow_length_ratio=0.3, color='black')

# 显示图形
plt.show()
```
